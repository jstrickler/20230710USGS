### Sum rows and column
Add a new column with sums of rows, and a new row with sums of columns

```python
df.loc['ALL ROWS'] = df.sum(numeric_only=True)
df['ALL COLUMNS'] = df.sum(numeric_only=True, axis=1)
df.tail()
```
