```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
import seaborn
seaborn.set()
```

```python
data = np.random.random(25)
data
data2 = np.random.random(25)
data2
```

```python
plt.subplots_adjust(hspace=.5, wspace=.5)
plt.subplot(2, 3, 1)
plt.title("plt.subplot(2, 3, 1)")
plt.plot(data, 2 * data)
plt.subplot(2, 3, 2)
plt.plot(data, 3 * data)
plt.title("plt.subplot(2, 3, 2)")
plt.subplot(2, 3, 3)
plt.title("plt.subplot(2, 3, 3)")
plt.plot(data, 10 * data)
plt.subplot(2, 3, 4)
plt.title("plt.subplot(2, 3, 4)")
plt.pie(data[:8])
plt.subplot(2, 3, 5)
plt.title("plt.subplot(2, 3, 5)")
plt.bar(sorted(data[:5]), data2[:5], .1)
plt.subplot(2, 3, 6)
plt.title("plt.subplot(2, 3, 6)")
plt.scatter(data, data2)
```
